```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
```

```python
df.head()
```

```python
df0 = df.drop('SkinThickness', axis='columns')
```

```python
df0
```

```python
df_x = df0.iloc[:, :-1]
df_x
```

```python
df_y = df0.iloc[:, 7]
df_y
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.25, random_state=21 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(random_state=16)
```

```python
LR.fit(x_train, y_train)
```

```python
pred = LR.predict(x_test)
```

```python
pred
```

```python
pred == y_test
```

```python
from sklearn import metrics
```

```python
cnf = metrics.confusion_matrix(y_test, pred)
```
